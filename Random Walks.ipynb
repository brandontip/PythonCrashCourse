```python
from random import choice

class RandomWalk:
    """Lets create some random walks!!"""
    
    def __init__(self, num_points=5000):
        self.num_points= num_points
        
        self.xcoords = [0]
        self.ycoords = [0]
    
    
    def fill_walk(self):
        while len(self.xcoords) < self.num_points:
            
            x_step = choice([0,1,2,3,4])*choice([-1,1])
            y_step = choice([0,1,2,3,4])*choice([-1,1])
            
            if x_step == 0 and y_step == 0:
                continue
            
            newx = self.xcoords[-1] + x_step
            newy = self.ycoords[-1] + y_step
            
            self.xcoords.append(newx)
            self.ycoords.append(newy)
            
```

```python
import matplotlib.pyplot as plt

rw = RandomWalk()
rw.fill_walk()

plt.style.use('classic')
#lets fill the screen
fig, ax = plt.subplots(figsize=(10,6), dpi=128)
point_numbers = range(rw.num_points)
ax.scatter(rw.xcoords, rw.ycoords, s=1, c=point_numbers, edgecolors='none' ,cmap=plt.cm.Blues)

ax.scatter(rw.xcoords[0], rw.ycoords[0], s=10, c='green', edgecolors='none' ,cmap=plt.cm.Blues)
ax.scatter(rw.xcoords[-1], rw.ycoords[-1], s=10, c='red', edgecolors='none' ,cmap=plt.cm.Blues)

ax.set_title("Random Walk w/ 5000 Steps ", fontsize=24)

plt.show()

```
