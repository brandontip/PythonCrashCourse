```python
import matplotlib.pyplot as plt
input_value = [1,2,3,4,5,6]
squares = [1,4,9,16,25,36]


fig, ax = plt.subplots()
ax.plot(input_value,squares)

plt.show()
```

```python
input_value = [1,2,3,4,5,6]
squares = [1,4,9,16,25,36]

fig, ax = plt.subplots()
ax.plot(input_value,squares, linewidth=3)
#Labeling
ax.set_title("Squares", fontsize=24)
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y=x^2", fontsize=14)

#tick labels
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
input_value = [1,2,3,4,5,6]
squares = [1,4,9,16,25,36]

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots()
ax.scatter(input_value,squares)
#Labeling
ax.set_title("Squares", fontsize=24)
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y=x^2", fontsize=14)

#tick labels
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
xs = range(1,1000)
ys = [x**2 for x in xs]

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots()
# s determines point size
# you can customize color too
ax.scatter(xs,ys, s=2)
#Labeling
ax.set_title("Squares", fontsize=24)
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y=x^2", fontsize=14)

#tick labels
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
xs = range(1,1000)
ys = [x**2 for x in xs]

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots()
# s determines point size
#c colormaps
ax.scatter(xs,ys, c=ys, cmap=plt.cm.Blues, s=2)
#Labeling
ax.set_title("Squares", fontsize=24)
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y=x^2", fontsize=14)

#tick labels
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

can save figures too

```python
xs = range(1,1000)
ys = [x**2 for x in xs]

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots()
# s determines point size
#c colormaps
ax.scatter(xs,ys, c=ys, cmap=plt.cm.Blues, s=2)
#Labeling
ax.set_title("Squares", fontsize=24)
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y=x^2", fontsize=14)

#tick labels
ax.tick_params(axis='both', labelsize=14)

# second arg trims whitespace
plt.savefig('squares_plot.png', bbox_inches='tight')
```
